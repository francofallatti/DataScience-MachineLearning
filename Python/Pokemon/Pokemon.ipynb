# Tipos de Pokemon y Gráfico de Pastel

Mostrar cuantos pokemons hay de cada tipo

```python
from collections import Counter

c= Counter(list(zip(gen1['type1'])))
type(c)
```

```python
typesTraduccion = ['Planta', 'Fuego', 'Agua','Insecto','Normal','Venenoso','Eléctrico','Tierra','Hada','Lucha','Psíquico','Roca','Fantasma','Fuego','Dragón']
```

Creamos un DataFrame con esta información

```python
lista = [list(i) for i in c.items()]
types_pokemon = pd.DataFrame(lista, index = typesTraduccion, columns=['tipo','cantidad'])
```

Ordenamos de mayor a menor

```python
types_pokemon = types_pokemon.sort_values(by=['cantidad'], ascending=False)
types_pokemon
```

```python
GraficaPie = types_pokemon.plot.pie(y='cantidad', figsize=(16, 10), autopct='%1.2f%%')
GraficaPie.get_legend().remove()
```

```python
Graficapie = types_pokemon.plot.pie(y="cantidad", title= 'Tipos de Pokemon',figsize=(16, 10),
           autopct='%1.2f%%',labels=['','','','','','','','','','','','','','',''])

Graficapie.legend(typesTraduccion,loc='upper left',ncol=2)

plt.savefig('Pokemon grafica de tipos.jpg')
```
