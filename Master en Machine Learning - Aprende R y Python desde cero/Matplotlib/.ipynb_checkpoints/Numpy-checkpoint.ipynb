# Numpy

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(212)
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.title("uso de subplot y figure")
plt.show()
```
